```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

import seaborn as sns
sns.set_style('whitegrid')
sinplot()

plt.show()
```
